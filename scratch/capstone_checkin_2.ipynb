# Capstone Project - Race Bib Identifier

### Problem Statement
For small races where chip timing is not utizilized, obtaining the rank order of finishers can be an onerous task.  After the finish line, the athletes are funneled into a single file line.  I will develope a model that is able to identify the bib numbers in a way that demonstrates its viability for use with a live video feed, and the ability to rank the athletes in order of bib number identification.

```python
import scipy.io as sio
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./Data/RBNR/set1_org/set1_32.JPG')
```

```python
RGB_im = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(RGB_im)
```

```python
mat_contents = sio.loadmat('./Data/RBNR/set1_org/set1_32.JPG.mat')
```

```python
mat_contents
```

```python
boxes = mat_contents['tagp']

for (minRow, maxRow, minCol, maxCol) in boxes:
    x = int(minCol)
    y = int(minRow)
    w = int(maxCol - minCol)
    h = int(maxRow - minRow)
    img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
```

```python
RGB_im = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_im)
```
